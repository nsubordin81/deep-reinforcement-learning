### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:

Please complete the function in the code cell below.

````python
""" this was my first crack at sarsa, took forever to run, the state value function 
has very high negative numbers instead of converging to something like the optimal.
I'm not sure what is wrong yet, noticed the solution used a different epsilong decay
but we both should have arguably converged in the 5000 episodes that were run. 
it seems like my q isn't being updated. have to keep looking at that."""

# taking these from the solution to monte carlo and modifying them
def sarsa_loop(env, Q, epsilon, alpha, nA):
    """ generates an episode from following the epsilon-greedy policy """
    state = env.reset()
    while True:
        # determine action from e-greedy policy
        action = epsilon_greedy_next_action(env, Q, state, epsilon, nA)
        # take action
        next_state, reward, done, info = env.step(action)
        next_action = epsilon_greedy_next_action(env, Q, next_state, epsilon, nA)
        # print(f"state: {state}, action: {action}, reward: {reward}, next_action: {next_action}, current Q at s,a: {Q[state][action]}")
        Q[state][action] += sarsa_update(Q, state, next_state, action, next_action, alpha, reward)
        # print(f"new state: {next_state}, new Q value {Q[state][action]}")
        state = next_state
        if done:
            break
    return Q

def epsilon_greedy_next_action(env, Q, state, epsilon, nA):
    return np.random.choice(np.arange(nA), p=get_probs(Q[state], epsilon, nA)) \
                                    if state in Q else env.action_space.sample()

# this is code from the course solution but now I get it, better python than I had
def get_probs(Q_s, epsilon, nA):
    """ obtains the action probabilities corresponding to epsilon-greedy policy """
    policy_s = np.ones(nA) * epsilon / nA
    best_a = np.argmax(Q_s)
    policy_s[best_a] = 1 - epsilon + (epsilon / nA)
    return policy_s

def sarsa_update(Q, s, s_prime, a, a_prime, alpha, reward):
    return alpha * (reward + Q[s_prime][a_prime] - Q[s][a])
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    eps_decay = .99999
    eps_min = 0.05
    epsilon = 1.0
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        # still doing this just like in monte carlo GLIE
        epsilon = max(epsilon*eps_decay, eps_min)
        # still getting a whole episode at a time, but need to update q inside this episode
        Q = sarsa_loop(env, Q, epsilon, alpha, env.nA)
    return Q
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 10, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````
